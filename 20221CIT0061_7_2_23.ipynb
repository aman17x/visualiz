```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,5,10)
print (x)
ax.plot(x,np.sin(x))
ax.set_title('simple plot')#adds title
ax.set_xlabel('wavelength')#adds label for x axis
ax.set_ylabel('amplitude')#adds label for y axis
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,100)
#print (x)
ax.plot(x,np.sin(x))
ax.plot(x,np.cos(x))
ax.set_title('Multiple lines')#adds title
ax.set_xlabel('wavelength')#adds label for x axis
ax.set_ylabel('amplitude')#adds label for y axis
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,100)
#print (x)
ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.cos(x),label='cos')#adding title for the corresponding graph
ax.set_title('Multiple lines')#adds title
ax.set_xlabel('wavelength')#adds label for x axis
ax.set_ylabel('amplitude')#adds label for y axis
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
#print (x)
ax.plot(x,np.sin(x),label='sin')
ax.plot(x,np.tan(x),label='tan')#adding title for the corresponding graph
ax.set_title('Multiple lines')#adds title
ax.set_xlabel('wavelength')#adds label for x axis
ax.set_ylabel('amplitude')#adds label for y axis
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
#print (x)
ax.plot(x,np.sin(x),label='sin',color='red')
ax.plot(x,np.cos(x),label='cos',color='cyan')#adding title for the corresponding graph
ax.set_title('Multiple lines')#adds title
ax.set_xlabel('wavelength')#adds label for x axis
ax.set_ylabel('amplitude')#adds label for y axis
ax.legend()
plt.show()
```

```python
fig=plt.figure()
ax=plt.axes()

x=np.linspace(0,10,1000)
#print (x)
ax.plot(x,np.sin(x),label='sin',linestyle='dashed')
ax.plot(x,np.cos(x),label='cos',linestyle='dotted')#adding title for the corresponding graph
ax.plot(x,np.sin(x+1),label='sin X+1',linestyle='dashdot')
ax.set_title('Multiple lines')#adds title
ax.set_xlabel('wavelength')#adds label for x axis
ax.set_ylabel('amplitude')#adds label for y axis
ax.legend()
plt.show()
```
